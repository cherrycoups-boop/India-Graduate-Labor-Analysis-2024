```python
print("t-test analysis of gender disparity in India's graduate workforce 2023-24:")
print("null hypothesis:There is no significant difference between male and female graduate participation in workforce")
print("alternate hypothesis:There is a significant difference between male and female participation in workforce")
from scipy import stats
import pandas as pd

df=pd.read_excel('india_graduate_workforce_genderdisparity_analysis (2).xlsx')


print("DataFrame columns:", df.columns.tolist())
print("DataFrame head:")
print(df.head())
print(df.shape)
print(f"number of rows:{df.shape[0]}")


t_stat, p_val=stats.ttest_rel(df['male graduate LFPR'], df['female graduate LFPR'])
print(f"t-statistic:{t_stat:.5f}")
print(f"p-value:{p_val:.4e}")
if p_val<0.05:
  print("Result:significant. There is a significant difference between male and female participation in workforce")
else:
  print("Result:not significant. There is no significant difference between male and female participation in workforce")
```
